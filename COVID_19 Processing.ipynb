```python
import pandas as pd
```

```python
file_path = './time_series_covid19_confirmed_US.csv'
data = pd.read_csv(file_path)
data
```

```python
list(data.columns)
```

```python
data['Combined_Key']
```

```python
cols = ['Province_State', 'Admin2', 'Lat', 'Long_'] + list(data.columns[11:])
data = data[cols]
data
```

```python
state_data = data.groupby(['Province_State','Admin2','Lat','Long_']).sum().reset_index()
state_data
```

```python
melted_data = state_data.melt(id_vars=['Province_State', 'Admin2','Lat', 'Long_'], var_name='Date', value_name='Confirmed')
melted_data
```

```python
melted_data['Date'] = pd.to_datetime(melted_data['Date'], format="%m/%d/%y")
melted_data
```

```python
melted_data.to_csv('processed_US_COVID_data.csv', index=False)
```

```python
agg_melted_data = melted_data.groupby(['Province_State','Date']).agg({'Confirmed':'sum'}).reset_index()
agg_melted_data
```

```python
agg_melted_data.to_csv('US_States_COVID_data.csv', index=False)
```
